```python
from tkinter import *
expression = ""

def press(num):
    global expression
    expression = expression + str(num)
    equation.set(expression)

def equalpress():
    try:
        global expression
        total = str(eval(expression))
        equation.set(total)
        expression = ""
    except:
        equation.set("Error")
        expression = ""

def clear():
    global expression
    expression = ""
    equation.set(0)

if __name__ == "__main__":
    root = Tk()
    root.title("Calculator")
    root.geometry("350x270")
    equation = StringVar()
    expression_field = Entry(root, textvariable = equation)
    expression_field.grid(columnspan=4, ipadx=70)
    clrscr = Button(root, text = 'C', fg = 'black', bg = 'grey', command = clear, height = 1, width = 7)
    clrscr.grid(row=1, column=0)
    square_root = Button(root, text = '√', fg = 'black', bg = 'grey', command = lambda: press('√'), height = 1, width = 7)
    square_root.grid(row=1, column=1)
    power = Button(root, text = 'x^y', fg = 'black', bg = 'grey', command = lambda: press('^'), height = 1, width = 7)
    power.grid(row=1, column=2)
    percentage = Button(root, text = '%', fg = 'black', bg = 'grey', command = lambda: press('%'), height = 1, width = 7)
    percentage.grid(row=1, column=3)
    num1 = Button(root, text = '1', fg = 'black', bg = 'grey', command = lambda: press('1'), height = 1, width = 7)
    num1.grid(row=2, column=0)
    num2 = Button(root, text = '2', fg = 'black', bg = 'grey', command = lambda: press('2'), height = 1, width = 7)
    num2.grid(row=2, column=1)
    num3 = Button(root, text = '3', fg = 'black', bg = 'grey', command = lambda: press('3'), height = 1, width = 7)
    num3.grid(row=2, column=2)
    num4 = Button(root, text = '4', fg = 'black', bg = 'grey', command = lambda: press('4'), height = 1, width = 7)
    num4.grid(row=3, column=0)
    num5 = Button(root, text = '5', fg = 'black', bg = 'grey', command = lambda: press('5'), height = 1, width = 7)
    num5.grid(row=3, column=1)
    num6 = Button(root, text = '6', fg = 'black', bg = 'grey', command = lambda: press('6'), height = 1, width = 7)
    num6.grid(row=3, column=2)
    num7 = Button(root, text = '7', fg = 'black', bg = 'grey', command = lambda: press('7'), height = 1, width = 7)
    num7.grid(row=4, column=0)
    num8 = Button(root, text = '8', fg = 'black', bg = 'grey', command = lambda: press('8'), height = 1, width = 7)
    num8.grid(row=4, column=1)
    num9 = Button(root, text = '9', fg = 'black', bg = 'grey', command = lambda: press('9'), height = 1, width = 7)
    num9.grid(row=4, column=2)
    num0 = Button(root, text = '0', fg = 'black', bg = 'grey', command = lambda: press('0'), height = 1, width = 7)
    num0.grid(row=5, column=0)
    plus = Button(root, text = '+', fg = 'black', bg = 'grey', command = lambda: press('+'), height = 1, width = 7)
    plus.grid(row=2, column=3)
    minus = Button(root, text = '-', fg = 'black', bg = 'grey', command = lambda: press('-'), height = 1, width = 7)
    minus.grid(row=3, column=3)
    mul = Button(root, text = '*', fg = 'black', bg = 'grey', command = lambda: press('*'), height = 1, width = 7)
    mul.grid(row=4, column=3)
    div = Button(root, text = '/', fg = 'black', bg = 'grey', command = lambda: press('/'), height = 1, width = 7)
    div.grid(row=5, column=3)
    point = Button(root, text = '.', fg = 'black', bg = 'grey', command = lambda: press('.'), height = 1, width = 7)
    point.grid(row=5, column=1)
    equal = Button(root, text = '=', fg = 'black', bg = 'orange', command = equalpress, height = 1, width = 7)
    equal.grid(row=5, column=2)
    root.mainloop()
```
